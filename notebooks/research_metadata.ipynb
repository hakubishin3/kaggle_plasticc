### 測光赤方偏移と分光赤方偏移の違い

- 分光赤方偏移は正確であるが、暗い天体の値を取得することは難しい（分光できないので）
- 代わりに測光赤方偏移を使う場合がある。
- 特に分光できないほど暗い銀河の赤方偏移を推定できるので、遠方銀河の研究で威力を発揮する。
- よく使われる測光赤方偏移の導出手法として、対象銀河の測光結果と最も一致する既存の銀河の赤方偏移を与えるというものがある。
- 銀河のSED（スペクトル分布）にはライマンブレイクなどの特徴的な段差や折れ曲がりがあるため、うまく機能する。
- しかし、分光赤方偏移よりは誤差が大きい。SEDの特徴を取り違えてしまう致命的な誤りがどうしても残ってしまう。
http://astro-dic.jp/photometric-redshift/

```python
# 分光赤方偏移が0でない、つまり銀河系外ソースである対象を抽出
extragal = train_meta['hostgal_specz'] != 0.
g = sns.jointplot(train_meta['hostgal_specz'][extragal],
                  train_meta['hostgal_photoz'][extragal], kind='hex',
                  xlim=(-0.01, 3.01), ylim=(-0.01,3.01))

# 分光赤方偏移と測光赤方偏移の差が大きい対象を抽出
outliers = np.abs(train_meta['hostgal_specz'] - train_meta['hostgal_photoz']) > 0.1
fig = g.fig
fig.axes[0].scatter(train_meta['hostgal_specz'][outliers],\
                    train_meta['hostgal_photoz'][outliers], color='C3', alpha=0.05)
fig.tight_layout()
plt.show()

# 差分のヒストグラム
plt.close()
sub = np.abs(train_meta['hostgal_specz'] - train_meta['hostgal_photoz'])
plt.hist(sub, bins=np.arange(0, 3, 0.1))
plt.xlabel('difference specz and photoz')
plt.show()
```

```python
# 分光赤方偏移が0でない、つまり銀河系外ソースである対象を抽出
extragal = test_meta['hostgal_specz'] != 0.
g = sns.jointplot(test_meta['hostgal_specz'][extragal],
                  test_meta['hostgal_photoz'][extragal], kind='hex',
                  xlim=(-0.01, 3.01), ylim=(-0.01,3.01))

# 分光赤方偏移と測光赤方偏移の差が大きい対象を抽出
outliers = np.abs(test_meta['hostgal_specz'] - test_meta['hostgal_photoz']) > 0.1
fig = g.fig
fig.axes[0].scatter(test_meta['hostgal_specz'][outliers],\
                    test_meta['hostgal_photoz'][outliers], color='C3', alpha=0.05)
fig.tight_layout()
plt.show()

# 差分のヒストグラム
plt.close()
sub = np.abs(test_meta['hostgal_specz'] - test_meta['hostgal_photoz']).dropna()
plt.hist(sub, bins=np.arange(0, 3, 0.1))
plt.xlabel('difference specz and photoz')
plt.show()
```

```python
target_list = list(np.sort(train_meta.target.unique()))
fig, axes = plt.subplots(ncols=2, nrows=14, figsize=(12, 48))
plt.subplots_adjust(wspace=0.4, hspace=0.6)
for ax, target in zip(axes, target_list):
    train_meta.query('target==@target and ddf==0').plot.scatter(
        ax=ax[0], x='gal_l', y='gal_b', s=20, title=f'class {target}, Wide-Fast-Deep (WFD, ddf=0)',
        xlim=(min(train_meta.gal_l), max(train_meta.gal_l)),
        ylim=(min(train_meta.gal_b), max(train_meta.gal_b))
    )
    train_meta.query('target==@target and ddf==1').plot.scatter(
        ax=ax[1], x='gal_l', y='gal_b', s=20, title=f'class {target}, Deep Drilling Fields (DDF, ddf=1)',
        xlim=(min(train_meta.gal_l), max(train_meta.gal_l)),
        ylim=(min(train_meta.gal_b), max(train_meta.gal_b))
    )
```

- DDFのエリアを5つに分割してみる。
- クラスタリングで5つのクラスタに分割。

```python
total_meta = pd.concat([train_meta, test_meta], axis=0, sort=False)
train_idx = range(0, len(train_meta))
test_idx = range(len(train_meta), len(train_meta)+len(test_meta))
total_ddf = total_meta.query('ddf==1')

kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(total_ddf[['gal_l', 'gal_b']])
pred = kmeans.predict(total_ddf[['gal_l', 'gal_b']])

total_ddf.plot.scatter(x='gal_l', y='gal_b', c=pred, cmap='Accent', s=10, title='total')
```

```python
result_list = []
for p in range(5):
    gal_l = total_ddf[pred == p]["gal_l"]
    gal_b = total_ddf[pred == p]["gal_b"]
    result_list.append((p, min(gal_l), max(gal_l), min(gal_b), max(gal_b)))
    
clusters = pd.DataFrame(result_list, columns=['cluster', 'min_gal_l', 'max_gal_l', 'min_gal_b', 'max_gal_b'])   
clusters
```

```python
fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(7, 5))
total_ddf.plot.scatter(ax=axes, x='gal_l', y='gal_b', c=pred, cmap='Accent', s=10, title='total')

for p in range(5):
    cluster = clusters.query('cluster==@p')
    min_gal_l = cluster["min_gal_l"].values
    max_gal_l = cluster["max_gal_l"].values
    min_gal_b = cluster["min_gal_b"].values
    max_gal_b = cluster["max_gal_b"].values
    rect = plt.Rectangle(
        (min_gal_l, min_gal_b), max_gal_l-min_gal_l, max_gal_b-min_gal_b, alpha=1, fill=None)
    axes.add_patch(rect)
plt.show()
```

- 取得した5つのDDF領域の確認を行う。
- 必要最小限の領域設定で、全てのDDF天体を抽出できている。

```python
n_ddf = 0
for p in range(5):
    cluster = clusters.query('cluster==@p')
    min_gal_l = cluster["min_gal_l"].values
    max_gal_l = cluster["max_gal_l"].values
    min_gal_b = cluster["min_gal_b"].values
    max_gal_b = cluster["max_gal_b"].values
    extract = train_meta.query(
        'gal_l >= @min_gal_l and gal_l <= @max_gal_l and gal_b >= @min_gal_b and gal_b <= @max_gal_b')
    
    print(f'=== cluster{p} ===')
    display(extract.groupby('ddf').size())
    n_ddf += extract.query('ddf==1').shape[0]

print(n_ddf, train_meta.query('ddf==1').shape[0])
```
