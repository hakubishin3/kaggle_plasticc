```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import numpy as np
from IPython.display import display
```

### データの読み込み

```python
train_meta = pd.read_csv("../data/input/training_set_metadata.csv")
test_meta = pd.read_csv("../data/input/test_set_metadata.csv")
print(train_meta.shape, test_meta.shape)
display(train_meta.head())
display(test_meta.head())
```

```python
train_predz = pd.read_feather("../data/feature/Predictor_specz_train.ftr")
test_predz = pd.read_feather("../data/feature/Predictor_specz_test.ftr")
display(train_predz.head())
display(test_predz.head())
```

```python
train_meta["predict_specz"] = train_predz
test_meta["predict_specz"] = test_predz
```

### 調査（train）

```python
# 分光赤方偏移が0でない、つまり銀河系外ソースである対象を抽出
extragal = train_meta['hostgal_photoz'] != 0.
g = sns.jointplot(train_meta['hostgal_specz'][extragal],
                  train_meta['hostgal_photoz'][extragal], kind='hex',
                  xlim=(-0.01, 3.01), ylim=(-0.01,3.01))

# 分光赤方偏移と測光赤方偏移の差が大きい対象を抽出
outliers = np.abs(train_meta['hostgal_specz'] - train_meta['hostgal_photoz']) > 0.1
fig = g.fig
fig.axes[0].scatter(train_meta['hostgal_specz'][outliers],\
                    train_meta['hostgal_photoz'][outliers], color='C3', alpha=0.05)
fig.tight_layout()
plt.show()

# 差分のヒストグラム
plt.close()
sub = np.abs(train_meta['hostgal_specz'] - train_meta['hostgal_photoz'])
plt.hist(sub, bins=np.arange(0, 3, 0.1))
plt.xlabel('difference specz and photoz')
plt.show()
```

```python
# 分光赤方偏移が0でない、つまり銀河系外ソースである対象を抽出
extragal = train_meta['hostgal_photoz'] != 0.
g = sns.jointplot(train_meta['hostgal_specz'][extragal],
                  train_meta['predict_specz'][extragal], kind='hex',
                  xlim=(-0.01, 3.01), ylim=(-0.01,3.01))

# 分光赤方偏移と測光赤方偏移の差が大きい対象を抽出
outliers = np.abs(train_meta['hostgal_specz'] - train_meta['predict_specz']) > 0.1
fig = g.fig
fig.axes[0].scatter(train_meta['hostgal_specz'][outliers],\
                    train_meta['predict_specz'][outliers], color='C3', alpha=0.05)
fig.tight_layout()
plt.show()

# 差分のヒストグラム
plt.close()
sub = np.abs(train_meta['hostgal_specz'] - train_meta['predict_specz'])
plt.hist(sub, bins=np.arange(0, 3, 0.1))
plt.xlabel('difference specz and photoz')
plt.show()
```

### 調査（test）

```python
# 分光赤方偏移が0でない、つまり銀河系外ソースである対象を抽出
extragal = test_meta['hostgal_photoz'] != 0.
g = sns.jointplot(test_meta['hostgal_specz'][extragal],
                  test_meta['hostgal_photoz'][extragal], kind='hex',
                  xlim=(-0.01, 3.01), ylim=(-0.01,3.01))

# 分光赤方偏移と測光赤方偏移の差が大きい対象を抽出
outliers = np.abs(test_meta['hostgal_specz'] - test_meta['hostgal_photoz']) > 0.1
fig = g.fig
fig.axes[0].scatter(test_meta['hostgal_specz'][outliers],\
                    test_meta['hostgal_photoz'][outliers], color='C3', alpha=0.05)
fig.tight_layout()
plt.show()

# 差分のヒストグラム
plt.close()
sub = np.abs(test_meta['hostgal_specz'] - test_meta['hostgal_photoz'])
plt.hist(sub[~np.isnan(sub)], bins=np.arange(0, 3, 0.1))
plt.xlabel('difference specz and photoz')
plt.show()
```
